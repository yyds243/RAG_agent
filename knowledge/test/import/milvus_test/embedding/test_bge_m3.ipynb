```python
from pymilvus.model.hybrid import BGEM3EmbeddingFunction

bge_m3_ef = BGEM3EmbeddingFunction(
    model_name=r'D:\A_model\bge-m3', # Specify the model name
    device='cuda:0', # Specify the device to use, e.g., 'cpu' or 'cuda:0'
    use_fp16=True # Specify whether to use fp16. Set to `False` if `device` is `cpu`.
)

texts = "我喜欢python"
embeddings = bge_m3_ef([texts])

# 打印出的就是你图片中对应的格式
print(embeddings)
```

```python
# indices (词表索引): [101, 50828, 992, 102]
#
# data (权重得分): [0.12, 0.45, 0.33, 0.11]
#
# indptr: [0, 4] (表示第一个向量用到了索引 0 到 4 的数据)

print(embeddings['dense'])
print(type(embeddings['dense'][0].tolist()))
print(type(embeddings['dense'][0]))
print(embeddings['dense'][0].tolist())
```

```python
start_index = embeddings['sparse'].indptr[0]
end_index = embeddings['sparse'].indptr[1]
weight = embeddings['sparse'].data[start_index:end_index].tolist()
tokenids = embeddings['sparse'].indices[start_index:end_index].tolist()
print(weight)
print(tokenids)
```

```python
sparse = dict(zip(tokenids, weight))
print(sparse)
```
